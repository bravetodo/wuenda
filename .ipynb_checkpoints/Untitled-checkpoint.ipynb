```python
x = [1, 2, 3]
y = [4, 5, 6]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=10,activation='relu',input_shape=[1]))
model.compile(optimizer='sgd',loss='mse')
model.fit(x,y, epochs=1000)
```

```python
model.predict([7])
```
